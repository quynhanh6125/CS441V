```python
df.set_index('Country', inplace=True)

# list các năm
years = list(range(1980, 2014))
```

**Question: Let's compare the number of immigrants from India and China from 1980 to 2013.**

```python
# lọc data của India và China
india = df.loc['India', years]
china = df.loc['China', years]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, india, label='India')
plt.plot(years, china, label='China')

plt.title('Immigration to Canada from India and China (1980–2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.legend()
plt.grid(True)

plt.show()
```

**Question: Compare the trend of top 5 countries that contributed the most to immigration to Canada.**

```python
# tìm 5 quốc gia hàng đầu theo tổng số người nhập cư
top5_countries = df[years].sum(axis=1).sort_values(ascending=False).head(5)

# trích xuất dữ liệu hàng năm cho 5 quốc gia hàng đầu
top5_data = df.loc[top5_countries.index, years].T

# Plot
plt.figure(figsize=(12, 7))
plt.plot(top5_data)

plt.title('Immigration Trend of Top 5 Source Countries (1980–2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.legend(top5_countries.index)
plt.grid(True)

plt.show()
```
