```python
df1 = df.assign(total_amount =df['quantity'] * df['unit_price'])
df1
```

```python
df2 = df1.query('total_amount > 500')
df2
```

```python
df_orders = df1.groupby('customer_id')['order_id'].count().reset_index(name = 'number of orders per customer')
df_orders
```

```python
df_spent = df1.groupby('customer_id')['total_amount'].sum().reset_index(name = 'total spent per customer')
df_spent
```

```python
df1['order_date'] = pd.to_datetime(df1['order_date'])
df1.info()
```

```python
df_revenue = df1.groupby('order_date')['total_amount'].sum().reset_index(name = 'revenue per day').sort_values('order_date')
df_revenue
```

```python
df_spent['rank'] = (df_spent['total spent per customer'].rank(method='dense', ascending=False))
df_spent
```

```python
df_spent = df_spent.sort_values('rank')
df_spent
```
